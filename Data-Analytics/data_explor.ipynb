```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
busyness_data= pd.read_csv('2020_Yellow_Taxi_Trip_Data_20240522.csv')

print(busyness_data.head())
```

## VendorID
- **Column Description**: A code indicating the TPEP provider that provided the record.
- **Expected/Allowed Values**: 1=Creative Mobile Technologies, LLC; 2=VeriFone Inc.
- **Field Limitations**: None
- **Additional Notes**: None

## tpep_pickup_datetime
- **Column Description**: The date and time when the meter was engaged.
- **Expected/Allowed Values**: None
- **Field Limitations**: None
- **Additional Notes**: None

## tpep_dropoff_datetime
- **Column Description**: The date and time when the meter was disengaged.
- **Expected/Allowed Values**: None
- **Field Limitations**: None
- **Additional Notes**: None

## passenger_count
- **Column Description**: The number of passengers in the vehicle.
- **Expected/Allowed Values**: None
- **Field Limitations**: None
- **Additional Notes**: This is a driver-entered value.

## trip_distance
- **Column Description**: The elapsed trip distance in miles reported by the taximeter.
- **Expected/Allowed Values**: None
- **Field Limitations**: None
- **Additional Notes**: None

## RatecodeID
- **Column Description**: The final rate code in effect at the end of the trip.
- **Expected/Allowed Values**: 1=Standard rate; 2=JFK; 3=Newark; 4=Nassau or Westchester; 5=Negotiated fare; 6=Group ride; 99=Null/unknown
- **Field Limitations**: None
- **Additional Notes**: None

## store_and_fwd_flag
- **Column Description**: This flag indicates whether the trip record was held in vehicle memory before sending to the vendor, aka “store and forward,” because the vehicle did not have a connection to the server.
- **Expected/Allowed Values**: Y=store and forward trip; N=not a store and forward trip
- **Field Limitations**: None
- **Additional Notes**: None

## PULocationID
- **Column Description**: TLC Taxi Zone in which the taximeter was engaged
- **Expected/Allowed Values**: None
- **Field Limitations**: None
- **Additional Notes**: None

## DOLocationID
- **Column Description**: TLC Taxi Zone in which the taximeter was disengaged
- **Expected/Allowed Values**: None
- **Field Limitations**: None
- **Additional Notes**: None

## payment_type
- **Column Description**: A numeric code signifying how the passenger paid for the trip.
- **Expected/Allowed Values**: 1=Credit card; 2=Cash; 3=No charge; 4=Dispute; 5=Unknown; 6=Voided trip
- **Field Limitations**: None
- **Additional Notes**: None

## fare_amount
- **Column Description**: The time-and-distance fare calculated by the meter.
- **Expected/Allowed Values**: None
- **Field Limitations**: None
- **Additional Notes**: None

## extra
- **Column Description**: Miscellaneous extras and surcharges. Currently, this only includes the $0.50 and $1 rush hour and overnight charges.
- **Expected/Allowed Values**: None
- **Field Limitations**: None
- **Additional Notes**: None

## mta_tax
- **Column Description**: $0.50 MTA tax that is automatically triggered based on the metered rate in use.
- **Expected/Allowed Values**: None
- **Field Limitations**: None
- **Additional Notes**: None

## tip_amount
- **Column Description**: Tip amount – This field is automatically populated for credit card tips. Cash tips are not included.
- **Expected/Allowed Values**: None
- **Field Limitations**: None
- **Additional Notes**: None

## tolls_amount
- **Column Description**: Total amount of all tolls paid during the trip.
- **Expected/Allowed Values**: None
- **Field Limitations**: None
- **Additional Notes**: None

## improvement_surcharge
- **Column Description**: $0.30 improvement surcharge assessed trips at the flag drop. The improvement surcharge began being levied in 2015.
- **Expected/Allowed Values**: None
- **Field Limitations**: None
- **Additional Notes**: None

## total_amount
- **Column Description**: The total amount charged to passengers. Does not include cash tips.
- **Expected/Allowed Values**: None
- **Field Limitations**: None
- **Additional Notes**: None

## congestion_surcharge
- **Column Description**: This surcharge is $2.50 and is assessed on trips that begin, end, or travel through Manhattan south of 96th Street. This went into effect on January 2019.
- **Expected/Allowed Values**: None
- **Field Limitations**: None
- **Additional Notes**: None

## airport_fee
- **Column Description**: $1.25 surcharge for trips to/from LaGuardia and JFK Airports. This went into effect in October 2020.
- **Expected/Allowed Values**: None
- **Field Limitations**: None
- **Additional Notes**: None

```python
import numpy as np 
import matplotlib.pyplot as plt
max_bins = 10 

for column in categorical_columns:
    # Here the data is being sorted
    count = busyness_data[column].value_counts().sort_index()
    total_count = count.sum()  # Total count of values
    categories_per_bin = np.ceil(len(count) / max_bins)

    if len(count) > max_bins:
        grouped_frequencies = []
        grouped_categories = []
        for i in range(0, len(count), int(categories_per_bin)):
            # labelling bins
            grouped_category = f'{count.index[i]} - {count.index[min(i + int(categories_per_bin) - 1, len(count) - 1)]}'
            grouped_categories.append(grouped_category)
            
            freq_sum = count.values[i:i + int(categories_per_bin)].sum()
            grouped_frequencies.append(freq_sum / total_count)  # Divide by total count to get proportion
        
        x_pos = np.arange(len(grouped_categories))
        frequencies = grouped_frequencies
        categories = grouped_categories
    else:
        #case where less than 10 bins are needed
        x_pos = np.arange(len(count))
        frequencies = count.values / total_count  # Divide by total count to get proportion
        categories = count.index

    plt.figure(figsize=(10, 6))
    plt.bar(x_pos, frequencies, align='center', alpha=0.7)
    plt.xticks(x_pos, categories, rotation=30, ha="right")
    plt.title(column)
    plt.tight_layout()
    plt.show()
```
